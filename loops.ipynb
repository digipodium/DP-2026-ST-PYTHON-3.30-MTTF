BREAK & CONTINUE

```python
lst = ['apple','mango','orange','tomato','kiwi','grapes']
for i in lst:
    if i == 'tomato':
        break
    print(i)
```

```python
lst = ['apple','mango','orange','tomato','kiwi','grapes']
for i in lst:
    if i == 'tomato':
        continue
    print(i)
```
