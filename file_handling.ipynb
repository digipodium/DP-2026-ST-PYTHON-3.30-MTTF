```python
#no need to close the file manually when using with statement
with open('data.txt','r') as file:
    content = file.read()
    print(content)
```

FILE PROPERTIES

```python
file = open('data.txt','r')
print(file.name)
print(file.mode)
print(file.closed)
print(file.tell()) #cursor position
file.read(5)
print(file.tell())
file.seek(8) #changes cursor position 
print(file.tell())
```

```python
try:
    file = open('newfile.txt','x')
    print("File created successfully.")
except FileExistsError:
    print("File already exists.")
finally:
    file.close()
```
