```python
import seaborn as sns
sns.get_dataset_names() #gives the list of available datasets in seaborn
```

```python
titanic = sns.load_dataset('titanic') #load the titanic dataset from seaborn
titanic.isnull().sum() #gives the count of null values in each column of the titanic dataset
```

- When a column has <40% missing values of the data, then we simply remove that  column
- When a column has >40% missing values of data, then we handle those missing values means we replace those missing values with any number
    - there are teo type of columns in any dataset - CATAGORICAL & NUMERICAL
    - When the column is numerical, we always repalce the null values with the column mean(average)
    - When the column is categorical, we always replace the null values with most frequent(mode) values 

```python
titanic.isnull().sum()
```

```python
titanic = titanic.dropna(subset=['embarked']) #drop the rows where the 'embarked' column has null values
titanic.isnull().sum() #gives the count of null values in each column
```

```python
avg_age = round(titanic['age'].mean()) #calculate the average age of passengers in the titanic dataset
print(avg_age)
```

```python
#replace null values with avg_age
titanic['age'] = titanic['age'].fillna(avg_age) #replace null values in the 'age' column with avg_age
titanic.isnull().sum()
```

```python
#checking duplicate values
duplicates = titanic[titanic.duplicated()]
print(duplicates)
```

```python
#dropping duplicates
titanic.drop_duplicates()
```

DATA MANIPULATION

```python
titanic
```

```python
#Adding a new column
titanic = titanic.assign(Discount=titanic["fare"]*0.10)
titanic
```

```python
titanic['fare'] = titanic['fare'].apply(lambda x:x*1.05)
titanic
```

```python
titanic["embarked"] = titanic["embarked"].map({
    "S" : "Southampton",
    "C" : "Cherbourg",
    "Q" : "Queenstown"
})
titanic
```

```python
titanic.groupby("embarked")["fare"].mean()
```

```python
titanic.groupby("sex")["fare"].max()
```
