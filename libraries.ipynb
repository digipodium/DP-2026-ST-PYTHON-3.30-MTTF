```python
from random import random, randint, choice, shuffle
```

```python
print("numbers between 0 and 1:", random())
print("numbers between given range:", randint(1, 10))
print("random choice from list:", choice([1, 2, 3, 4, 5]))
name = ["JOhn","Jane","Jack","Jill"]
shuffle(name)
print("shuffled list:", name)
```

```python
from datetime import datetime as dt #alias
```

```python
print("Current date and time:", dt.now())
print("Current date:", dt.now().date())
print("Current time:", dt.now().time())
print("Current year:", dt.now().year)
print("Current month:", dt.now().month)
print("Current day:", dt.now().day)
print("Current hour:", dt.now().hour)
print("Current minute:", dt.now().minute)
print("Current second:", dt.now().second)
print("Current microsecond:", dt.now().microsecond)
```

```python
print("Formatted Date:", dt.now().strftime("%d-%m-%Y"))
print("Formatted Time:", dt.now().strftime("%H:%M:%S"))
print("Current Month Name:", dt.now().strftime("%B"))
print("Current Month Name:", dt.now().strftime("%b"))
print("Current Day Name:", dt.now().strftime("%A"))
print("Current Day Name:", dt.now().strftime("%a"))
```

```python
total_files = 0
total_folders = 0
for path, folders, files in os.walk(os.getcwd()):
    print("Current Path:", path)
    print("Folders:", folders)
    print("Files:", files)
    total_files += len(files)
    total_folders += len(folders)
    
print("Total Files:", total_files)
print("Total Folders:", total_folders)
```
