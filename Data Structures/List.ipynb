- append : adds an element at the end of the list
- insert : add an element at any specific position
- pop : deletes an element using index value
- remove : deletes an element using the element name
- sort : arrange the elements in particular order
- copy : make a new copy of the list
- extend : combines two list
- clear : removes all the elements from the list
- reverse : reverse the list

```python
apps = ["google","drive","netflix","Watsap","blinkit","Snapchat"]
print(apps)
```

```python
apps.append('facebook')
print(apps)
```

```python
apps.insert(2, 'swiggy')
print(apps)
```

```python
apps.pop() #it will delete last element from the list
print(apps)
```

```python
apps.pop(2)
print(apps)
```

```python
apps.remove("drive")
print(apps)
```

```python
apps2 = apps.copy()
print(apps2)
```

```python
apps2.clear()
print(apps2)
```

```python
apps.sort()
print(apps)
```

```python
lst = [2,5,6,3,5]
apps.extend(lst)
print(apps)
```

```python
apps.reverse()
print(apps)
```
