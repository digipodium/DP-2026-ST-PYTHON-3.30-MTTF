```python
print(student.keys())
print(student.values())
print(student.items())
```

```python
for i,j in student.items():
    print(f"{i} : {j}")
```

```python
#adding a new key value pair
student['city'] = "Lko"
print(student)
```

```python
#deleting a key value pair
del student['age']
print(student)
```

```python
#updating a dictionary
student['name'] = "Raj"
print(student)
```
