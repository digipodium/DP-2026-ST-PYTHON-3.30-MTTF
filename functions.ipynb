`FUNCTION` : a block of code to reuse any functionality

`Types of functions`
- user defined functions
    - parameterised functions
    - non parameterised functions
    - return functions
    - non return functions
    - default parameterised functions
    - lambda functions
- built in functions
    - append, len etc.

```python
#non parameterised function
def greet(): #function definition
    print("Hello user")
    print("Welcome to my first function prog.")
    print("Have a nice day!")

greet()
print("See you soon")
greet()
```
