```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
df
```

```python
sns.histplot(df,x='total_bill',kde=True,color='red')
plt.title('Distribution of Total Bill')
plt.show()
```

```python
sns.countplot(df, x = 'day')
plt.title("Customers on Different Day")
plt.show()
```

```python
sns.scatterplot(df, x='total_bill',y='tip',hue='sex',style='time')
plt.title("Relationship between total bill and its tip given")
plt.show()
```

```python
sns.lineplot(df, x='size',y='tip',estimator='mean',marker='o')
plt.title('Average tip by Group size')
plt.show()
```

- genderwise count of per day
- bar plot of average bill by day

```python
sns.violinplot(df, x='day',y='tip',hue='sex',split=True,inner='quart')
plt.title("Gender and Day Wise Tip")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
